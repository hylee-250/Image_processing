```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
fd = open("C:\\Work\\Python\\Image_processing\\lena(512x512).raw",'rb')
rows = 512
cols = 512
f = np.fromfile(fd, dtype=np.uint8,count=rows*cols)
im = f.reshape((rows,cols))
cv2.imwrite("C:\\Work\\Python\\Image_processing\\lena(512x512).png",im)
cv2.imwrite("C:\\Work\\Python\\Image_processing\\lena(512x512).bmp",im)
cv2.imwrite("C:\\Work\\Python\\Image_processing\\lena(512x512).jpg",im)
fd.close()

imgJPG = cv2.imread("C:\\Work\\Python\\Image_processing\\lena(512x512).jpg")
imgPNG = cv2.imread("C:\\Work\\Python\\Image_processing\\lena(512x512).png")
imgBMP = cv2.imread("C:\\Work\\Python\\Image_processing\\lena(512x512).bmp")
fig = plt.figure(figsize = (16,12))
rows = 1
cols = 3
ax1 = fig.add_subplot(rows,cols,1)
ax1.imshow(cv2.cvtColor(imgBMP,cv2.COLOR_BGR2RGB))
ax1.set_xlabel('bmp')
ax1.set_xticks([]), ax1.set_yticks([])

ax2 = fig.add_subplot(rows,cols,2)
ax2.imshow(cv2.cvtColor(imgPNG,cv2.COLOR_BGR2RGB))
ax2.set_xlabel('png')
ax2.set_xticks([]), ax2.set_yticks([])

ax3 = fig.add_subplot(rows,cols,3)
ax3.imshow(cv2.cvtColor(imgJPG,cv2.COLOR_BGR2RGB))
ax3.set_xlabel('jpg')
ax3.set_xticks([]), ax3.set_yticks([])

plt.show()
cv2.waitKey()
cv2.destroyAllWindows()
```
