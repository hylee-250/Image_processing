```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_10 = cv2.imread("C:\\Work\\Python\\Image_processing\\10.jpg",cv2.IMREAD_COLOR)
img_50 = cv2.imread("C:\\Work\\Python\\Image_processing\\50.jpg",cv2.IMREAD_COLOR)
img_100 = cv2.imread("C:\\Work\\Python\\Image_processing\\100.jpg",cv2.IMREAD_COLOR)
img_500 = cv2.imread("C:\\Work\\Python\\Image_processing\\500.jpg",cv2.IMREAD_COLOR)
gray_10 = cv2.cvtColor(img_10,cv2.COLOR_RGB2GRAY).astype(np.double)
gray_50 = cv2.cvtColor(img_50,cv2.COLOR_RGB2GRAY).astype(np.double)
gray_100 = cv2.cvtColor(img_100,cv2.COLOR_RGB2GRAY).astype(np.double)
gray_500 = cv2.cvtColor(img_500,cv2.COLOR_RGB2GRAY).astype(np.double)
height, width = gray_50.shape

gray_10 = gray_10/255
gray_50 = gray_50/255
gray_100 = gray_100/255
gray_500 = gray_500/255


pad_10 = cv2.copyMakeBorder(gray_10,1,1,1,1,cv2.BORDER_REPLICATE)
pad_50 = cv2.copyMakeBorder(gray_50,1,1,1,1,cv2.BORDER_REPLICATE)
pad_100 = cv2.copyMakeBorder(gray_100,1,1,1,1,cv2.BORDER_REPLICATE)
pad_500 = cv2.copyMakeBorder(gray_500,1,1,1,1,cv2.BORDER_REPLICATE)

sobel_x_filter = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel_y_filter = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

edge_x_10 = np.zeros((height,width))
edge_y_10 = np.zeros((height,width))
edge_10 = np.zeros((height,width))

edge_x_50 = np.zeros((height,width))
edge_y_50 = np.zeros((height,width))
edge_50 = np.zeros((height,width))

edge_x_100 = np.zeros((height,width))
edge_y_100 = np.zeros((height,width))
edge_100 = np.zeros((height,width))

edge_x_500 = np.zeros((height,width))
edge_y_500 = np.zeros((height,width))
edge_500 = np.zeros((height,width))
```

```python
threshold = 0.3

for i in range(height):
    for j in range(width):
        # 50
        edge_x_10[i][j] = np.sum(pad_10[i:i+3,j:j+3] * sobel_x_filter)
        edge_y_10[i][j] = np.sum(pad_10[i:i+3,j:j+3] * sobel_y_filter)
        temp_10 = (np.abs(edge_x_10[i][j]) + np.abs(edge_y_10[i][j]))
        if(temp_10 < threshold):
            edge_10[i][j] = 0
        else:
            edge_10[i][j] = temp_10
        # 50
        edge_x_50[i][j] = np.sum(pad_50[i:i+3,j:j+3] * sobel_x_filter)
        edge_y_50[i][j] = np.sum(pad_50[i:i+3,j:j+3] * sobel_y_filter)
        temp_50 = (np.abs(edge_x_50[i][j]) + np.abs(edge_y_50[i][j]))
        if(temp_50 < threshold):
            edge_50[i][j] = 0
        else:
            edge_50[i][j] = temp_50
        # 100
        edge_x_100[i][j] = np.sum(pad_100[i:i+3,j:j+3] * sobel_x_filter)
        edge_y_100[i][j] = np.sum(pad_100[i:i+3,j:j+3] * sobel_y_filter)
        temp_100 = (abs(edge_x_100[i][j]) + abs(edge_y_100[i][j]))
        if(temp_100 < threshold):
            edge_100[i][j] = 0
        else:
            edge_100[i][j] = temp_100
        # 500
        edge_x_500[i][j] = np.sum(pad_500[i:i+3,j:j+3] * sobel_x_filter)
        edge_y_500[i][j] = np.sum(pad_500[i:i+3,j:j+3] * sobel_y_filter)
        temp_500 = (abs(edge_x_500[i][j]) + abs(edge_y_500[i][j]))
        if(temp_500 < threshold):
            edge_500[i][j] = 0
        else:
            edge_500[i][j] = temp_500
```

```python
rows = 3
cols = 3
fig, axes = plt.subplots(rows,cols,figsize=(10,10))


axes[0,0].imshow(edge_x_50,cmap='gray')
axes[0,0].set_title('edge x axis 50')
axes[0,0].set_xticks([])
axes[0,0].set_yticks([])

axes[0,1].imshow(edge_y_50,cmap='gray')
axes[0,1].set_title('edge y axis 50')
axes[0,1].set_xticks([])
axes[0,1].set_yticks([])

axes[0,2].imshow(edge_50,cmap='gray')
axes[0,2].set_title('edge 50')
axes[0,2].set_xticks([])
axes[0,2].set_yticks([])

axes[1,0].imshow(edge_x_100,cmap='gray')
axes[1,0].set_title('edge x axis 100')
axes[1,0].set_xticks([])
axes[1,0].set_yticks([])

axes[1,1].imshow(edge_y_100,cmap='gray')
axes[1,1].set_title('edge y axis 100')
axes[1,1].set_xticks([])
axes[1,1].set_yticks([])

axes[1,2].imshow(edge_100,cmap='gray')
axes[1,2].set_title('edge 100')
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

axes[2,0].imshow(edge_x_500,cmap='gray')
axes[2,0].set_title('edge x axis 500')
axes[2,0].set_xticks([])
axes[2,0].set_yticks([])

axes[2,1].imshow(edge_y_500,cmap='gray')
axes[2,1].set_title('edge y axis 500')
axes[2,1].set_xticks([])
axes[2,1].set_yticks([])

axes[2,2].imshow(edge_500,cmap='gray')
axes[2,2].set_title('edge 500')
axes[2,2].set_xticks([])
axes[2,2].set_yticks([])

plt.show()

rows = 1
cols = 3
fig, axes = plt.subplots(rows,cols,figsize=(10,10))


axes[0].imshow(edge_x_10,cmap='gray')
axes[0].set_title('edge x axis 10')
axes[0].set_xticks([])
axes[0].set_yticks([])

axes[1].imshow(edge_y_10,cmap='gray')
axes[1].set_title('edge y axis 10')
axes[1].set_xticks([])
axes[1].set_yticks([])

axes[2].imshow(edge_10,cmap='gray')
axes[2].set_title('edge 10')
axes[2].set_xticks([])
axes[2].set_yticks([])
```

```python
def diameter_check(img):
    height,width = img.shape
    left = width
    right = 0
    top = height
    bottom = 0
    for i in range(height):
        for j in range(width):
            if((img[i][j]!=0) &(j<left)):
                left = j
            if((img[i][j]!=0) &(j>right)):
                right = j
            if((img[i][j]!=0) &(i<top)):
                top = i
            if((img[i][j]!=0) &(i>bottom)):
                bottom = i
    diameter = np.mean([right-left,bottom-top])*21.6/220
    return diameter
```

```python
print(f"이 동전의 지름은 {diameter_check(edge_10):.3f}mm입니다.")
print(f"이 동전의 지름은 {diameter_check(edge_50):.3f}mm입니다.")
print(f"이 동전의 지름은 {diameter_check(edge_100):.3f}mm입니다.")
print(f"이 동전의 지름은 {diameter_check(edge_500):.3f}mm입니다.")
```

```python
def coin_classifier(img):
    size_50 = 21.6
    size_100 = 24
    size_500 = 26.5
    if((diameter_check(img)>=0.9*size_50) & (diameter_check(img)<=1.1*size_50)):
        print("50원")
    elif((diameter_check(img)>=0.9*size_100) & (diameter_check(img)<=1.1*size_100)):
        print("100원")
    elif((diameter_check(img)>=0.9*size_500) & (diameter_check(img)<=1.1*size_500)):
        print("500원")
```

```python
coin_classifier(edge_50)
coin_classifier(edge_100)
coin_classifier(edge_500)
```
