```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def LPF50(img):
    rows, cols = img.shape
    lpf_50 = np.zeros((rows,cols))
    for i in range(rows):
        for j in range(cols):
            if((i >=230) & (i<280)):
                if((j >=230) & (j<280)):
                    lpf_50[i][j] = 1
    return lpf_50

def LPF100(img):
    rows, cols = img.shape
    lpf_100 = np.zeros((rows,cols))
    for i in range(rows):
        for j in range(cols):
            if((i >=205) & (i<305)):
                if((j >=205) &(j<305)):
                    lpf_100[i][j] = 1
    return lpf_100

def LPF200(img):
    rows, cols = img.shape
    lpf_200 = np.zeros((rows,cols))
    for i in range(rows):
        for j in range(cols):
            if((i >=155) & (i<305)):
                if((j >=155) & (j<305)):
                    lpf_200[i][j] = 1
    return lpf_200

def LPF400(img):
    rows, cols = img.shape
    lpf_400 = np.zeros((rows,cols))
    for i in range(rows):
        for j in range(cols):
            if((i >=55) & (i<455)):
                if((j >=55) & (j<455)):
                    lpf_400[i][j] = 1
    return lpf_400
```

```python
tiff_img = cv2.imread("C:\\Work\\Python\\Image_processing\\lena.tiff",cv2.IMREAD_COLOR)
gray_img = cv2.cvtColor(tiff_img,cv2.COLOR_RGB2GRAY)
height, width = gray_img.shape

fft_img = np.fft.fft2(gray_img)
fft_shift = np.fft.fftshift(fft_img)
lowpass_50 = LPF50(gray_img)
lowpass_100 = LPF100(gray_img)
lowpass_200 = LPF200(gray_img)
lowpass_400 = LPF400(gray_img)


filtered_50_img = fft_shift * lowpass_50
filtered_100_img = fft_shift * lowpass_100
filtered_200_img = fft_shift * lowpass_200
filtered_400_img = fft_shift * lowpass_400

ifft_img_50 = abs(np.fft.ifft2(filtered_50_img)).astype(np.uint8)
ifft_img_100 = abs(np.fft.ifft2(filtered_100_img)).astype(np.uint8)
ifft_img_200 = abs(np.fft.ifft2(filtered_200_img)).astype(np.uint8)
ifft_img_400 = abs(np.fft.ifft2(filtered_400_img)).astype(np.uint8)

#print(np.sum(lowpass_50))
#print(np.sum(lowpass_100))
#print(np.sum(lowpass_200))
#print(np.sum(lowpass_400))

#print(ifft_img_50)
rows = 2
cols = 2
fig, axes = plt.subplots(rows,cols,figsize=(10,10))

axes[0,0].imshow(ifft_img_50,cmap='gray')
axes[0,0].set_title('Lowpass 50')
axes[0,0].set_xticks([])
axes[0,0].set_yticks([])

axes[0,1].imshow(ifft_img_100,cmap='gray')
axes[0,1].set_title('Lowpass 100')
axes[0,1].set_xticks([])
axes[0,1].set_yticks([])

axes[1,0].imshow(ifft_img_200,cmap='gray')
axes[1,0].set_title('Lowpass 200')
axes[1,0].set_xticks([])
axes[1,0].set_yticks([])

axes[1,1].imshow(ifft_img_400,cmap='gray')
axes[1,1].set_title('Lowpass 400')
axes[1,1].set_xticks([])
axes[1,1].set_yticks([])


plt.show()
```
