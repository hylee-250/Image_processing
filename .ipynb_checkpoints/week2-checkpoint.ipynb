```python
# tiff  ,    
tiff_img = cv2.imread("C:\\Work\\Python\\Image_processing\\lena.tiff",cv2.IMREAD_COLOR)
gray_img = cv2.cvtColor(tiff_img,cv2.COLOR_RGB2GRAY)

#     1/4    
dst = np.clip((gray_img/4),0,255).astype(np.uint8)

#     
hist = np.zeros(256)
for i in range(512):
    for j in range(512):
        k = dst[i][j]
        hist[k] += 1
        
#  ,    pdf,   256  cdf 
pdf = hist/(512*512)
cdf = np.cumsum(pdf)*256

#   
sum_hist = np.cumsum(hist)

# normalize   equalize  
equalize = np.zeros(256)
for i in range(256):
    equalize[i]=round(sum_hist[i]*255/(512 **2)+0.5)
restored = np.zeros((512,512))

#   equalize    .
for i in range(512):
    for j in range(512):
        restored[i][j] = equalize[dst[i][j]]

restored = restored.astype(np.uint8)

fig = plt.figure(figsize=(12,6))
rows = 1
cols = 4
ax1 = fig.add_subplot(rows,cols,1)
ax1.hist(dst.ravel(),256,[0,256])
ax1.set_xlabel('dark img')

ax2 = fig.add_subplot(rows,cols,2)
ax2.hist(x =[i for i in range(256)],bins=256,weights= pdf.reshape(-1))
ax2.set_xlabel('pdf')

ax3 = fig.add_subplot(rows,cols,3)
ax3.hist(x =[i for i in range(256)],bins=256,weights= cdf.reshape(-1))
ax3.set_xlabel('cdf')

ax4 = fig.add_subplot(rows,cols,4)
ax4.hist(restored.ravel(),256,[0,256])
ax4.set_xlabel('restored')
plt.show()

fig = plt.figure(figsize = (16,12))
rows = 1
cols = 3

ax1 = fig.add_subplot(rows,cols,1)
ax1.imshow(gray_img,cmap='gray',vmin=0,vmax=255)
ax1.set_xlabel('original')
ax1.set_xticks([]), ax1.set_yticks([])


ax2 = fig.add_subplot(rows,cols,2)
ax2.imshow(dst,cmap='gray',vmin=0,vmax=255)
ax2.set_xlabel('dark')
ax2.set_xticks([]), ax2.set_yticks([])

ax3 = fig.add_subplot(rows,cols,3)
ax3.imshow(restored,cmap='gray',vmin=0,vmax=255)
ax3.set_xlabel('restored')
ax3.set_xticks([]), ax3.set_yticks([])

plt.show()
cv2.waitKey()
cv2.destroyAllWindows()
```
